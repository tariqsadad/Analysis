```python
df = pd.read_csv('./nslkdd/kdd_train.csv')  # Adjust the path as needed
df.head(5)
```

```python
# Separate features and labels
X = df.iloc[:, :-1]  # Features
y = df.iloc[:, -1]   # Labels
```

```python
print("Actual columns in X:", X.columns.tolist())  # Debug step
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, RobustScaler, OneHotEncoder

# Convert X to a DataFrame (if it isn't already)
X = pd.DataFrame(X)

# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Encode categorical features using OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(X[categorical_cols])

# Convert remaining numeric columns to a NumPy array
numeric_cols = X.select_dtypes(include=['number']).values

# Concatenate encoded categorical features with numeric features
X_processed = np.hstack((X_encoded, numeric_cols))

```

```python
# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Normalize using RobustScaler
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X_processed)

```

```python
# K-Means Outlier Detection
# ----------------------

# Elbow Method for optimal K
sse = []
k_range = range(2, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()

# Apply K-Means (k=5 based on elbow plot)
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Remove outliers (top 5% distance)
distances = kmeans.transform(X_scaled)
outlier_threshold = np.percentile(distances.min(axis=1), 95)
outliers = distances.min(axis=1) > outlier_threshold
X_filtered = X_scaled[~outliers]
y_filtered = y_encoded[~outliers]
```

```python
# Dimensionality Reduction
# ----------------------
pca = PCA(n_components=0.95)  # Preserve 95% variance
X_pca = pca.fit_transform(X_filtered)
```

```python
# ----------------------
# LSTM Model Preparation
# ----------------------

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y_filtered,
    test_size=0.2,
    random_state=42
)

# Reshape for LSTM (samples, timesteps, features)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# One-hot encode labels
num_classes = len(np.unique(y_filtered))
y_train = to_categorical(y_train, num_classes=num_classes)
y_test = to_categorical(y_test, num_classes=num_classes)

# ----------------------
# LSTM Model Architecture
# ----------------------
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ----------------------
# Model Training
# ----------------------
start_time = time.time()
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)
training_time = time.time() - start_time

# ----------------------
# Evaluation & Visualization
# ----------------------

# Generate predictions
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Classification Report
print("Classification Report:")
print(classification_report(y_test_classes, y_pred_classes))

# Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test_classes, y_pred_classes),
            annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curves
plt.figure(figsize=(10, 8))
for i in range(num_classes):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')
plt.title('ROC Curves')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# Training History
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Accuracy Evolution')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Loss Evolution')
plt.ylabel('Loss')
plt.legend()
plt.show()

# ----------------------
# Computational Analysis
# ----------------------
print(f"\nTraining Time: {training_time:.2f} seconds")
print(f"Model Parameters: {model.count_params()}")
print(f"Final Validation Accuracy: {np.max(history.history['val_accuracy']):.2%}")

# Confidence Interval Calculation
accuracy = np.mean(y_pred_classes == y_test_classes)
n = len(y_test_classes)
std_err = np.sqrt(accuracy * (1 - accuracy) / n)
ci = 1.96 * std_err
print(f"Accuracy: {accuracy:.2%} (±{ci:.2%} 95% CI)")

# Save model
model.save('ids_model.h5')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score

# Load your dataset
# X = pd.read_csv('your_data.csv')

# --- Data Preprocessing ---
# 1. Convert numeric columns (handle string-formatted numbers)
numeric_cols = X.select_dtypes(include=[np.number]).columns
X[numeric_cols] = X[numeric_cols].apply(pd.to_numeric, errors='coerce')

# 2. Handle missing values
X = X.dropna()

# 3. Encode categorical features
categorical_cols = X.select_dtypes(include=['object', 'category']).columns
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_cat_encoded = encoder.fit_transform(X[categorical_cols])
X_processed = np.hstack([X[numeric_cols].values, X_cat_encoded])

# --- Dimensionality Reduction ---
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_processed)

# --- Clustering with K=5 ---
kmeans = MiniBatchKMeans(n_clusters=5,
                        batch_size=1024,
                        random_state=42,
                        n_init=10)
clusters = kmeans.fit_predict(X_pca)

# --- Visualization ---
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1],
                     c=clusters, cmap='viridis',
                     s=15, alpha=0.7, edgecolor='k')
centers = plt.scatter(kmeans.cluster_centers_[:, 0],
                     kmeans.cluster_centers_[:, 1],
                     c='red', marker='X', s=200,
                     edgecolor='black', linewidth=1)

# Add variance explained to labels
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% Variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% Variance)')
plt.title('K-Means Clustering with K=5')
plt.colorbar(scatter, label='Cluster ID')
plt.grid(alpha=0.3)

# Add cluster size annotations
cluster_counts = np.bincount(clusters)
for i, count in enumerate(cluster_counts):
    plt.text(kmeans.cluster_centers_[i, 0],
            kmeans.cluster_centers_[i, 1],
            f'Size: {count}\nID: {i}',
            ha='center', va='center',
            bbox=dict(facecolor='white', alpha=0.8))

plt.show()

# --- Cluster Analysis ---
print(f"Silhouette Score: {silhouette_score(X_pca, clusters):.3f}")
print("\nCluster Distribution:")
print(pd.Series(clusters).value_counts().sort_index())
```
